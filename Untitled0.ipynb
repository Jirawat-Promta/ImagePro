```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
from skimage.util import img_as_ubyte
from google.colab.patches import cv2_imshow
```

```python
orginput_img = cv2.imread("/content/bubble_in_cement.jpg")
#cv2_imshow(orginput_img)
#print(orginput_img.shape)

r = int(input("Row in img = "))
#print(r)

#ml =  np.zeros(orginput_img.shape[:2],np.uint8)
ml = np.zeros((orginput_img.shape[0],orginput_img.shape[1]),np.uint8)
#print(ml.shape)
#cv2.imshow(ml)
cv2.line(ml,(0,r),(ml.shape[1],r),(255,255,255),1)

hsv_img = cv2.cvtColor(orginput_img,cv2.COLOR_BGR2HSV)
lower = np.array([60,100,100])
upper = np.array([120,255,255])
bb_All = cv2.inRange(hsv_img,lower,upper)

contours,hierarchys = cv2.findContours(bb_All,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
mb = np.zeros(orginput_img.shape[:2],np.uint8)

for i in range(len(contours)) :
  if cv2.contourArea(contours[i])>500 :
    cv2.drawContours(mb,contours,i,(255,255,255),-1)
    cv2.drawContours(orginput_img,contours,i,(0,0,255),2)
bbm = ml & mb

pr = bbm[r , 0]
count_bb = 0
START= [ ]
STOP = [ ]
for i in range(bbm.shape[1]) :
    if pr != bbm[r,i] :
      count_bb = count_bb + 1
      if count_bb%2  != 0 :
        START.append(i)
      else :
        STOP.append(i)
    pr = bbm [r, i ]

#print(START)
#print(STOP)

dpmm = 600/25.4
w = []
y = [ ]
x = r/dpmm
for i in range(len(START)) :
  w.append((STOP[i] - START[i])/dpmm)
  y.append((((STOP[i] - START[i])/2)+START[i])/dpmm)

#แสดงผล
print("\n n = ",len(START))

for i in range(len(START)) :
  print("bubble",str(i+1))
  print("width ={:.2f}".format(w[i]),"mm." )
  print("(x,y) = ({:.2f},".format(x) +"{:.2f})".format(y[i]),"mm.")

cv2.line(orginput_img,(0,r),(ml.shape[1],r),(200,100,255),2)
cv2.putText(orginput_img,("n = "+str(len(START))),(5,(r-400)),cv2.FONT_HERSHEY_DUPLEX,1,(255,255,255),2)
for i in range(len(START)) :
  cv2.putText(orginput_img,("bubble = "+str(i+1)),(5,(r-300)+(i*170)),cv2.FONT_HERSHEY_DUPLEX,1,(255,255,255),2)
  cv2.putText(orginput_img,("width = {:.2f}".format(w[i])+"mm."),(5,(r-250)+(i*170)),cv2.FONT_HERSHEY_DUPLEX,1,(255,255,255),2)
  cv2.putText(orginput_img,("(x,y) = ({:.2f}, ".format(x)+"{:.2f})".format(y[i])+"mm."),(5,(r-200)+(i*170)),cv2.FONT_HERSHEY_DUPLEX,1,(255,255,255),2)

cv2_imshow(orginput_img)
```
